## What is Jupyter?

Jupyter is a web-based interactive computing environment. Each **notebook** is a document containing:
- **Markdown cells** for text, equations, and diagrams (like this one)
- **Code cells** for running Python code in the browser

All notebooks run in **JupyterLite**, which is a browser-only version of Jupyter that requires no server or installation. Every Python library is pre-loaded via Pyodide (a Python interpreter compiled to WebAssembly).

**What you can do:**
- Edit and run code cells with `Shift+Enter`
- Change parameters and re-run to see what happens
- Interact with sliders, dropdowns, and buttons powered by `ipywidgets`
- View plots, animations, and live-updating figures

## Interactive Example (Step Response)

Run the cell below and drag the sliders to see how the natural frequency $\omega_n$ and damping ratio $\zeta$ affect the step response of a second-order system. Please note that it is a demo to show you what interactive elements to expect from the course notebooks, and it is possible you have yet to learn what exactly this means. 

```python
%pip install -q ipywidgets

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import step, TransferFunction
import ipywidgets as widgets
from IPython.display import display

def plot_step(zeta, omega_n):
    sys = TransferFunction([omega_n**2], [1, 2*zeta*omega_n, omega_n**2])
    t, y = step(sys)

    fig, ax = plt.subplots(figsize=(8, 3.5))
    ax.plot(t, y, 'C0', lw=1.5)
    ax.axhline(1, color='gray', ls='--', lw=0.7, label='Reference')
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Output')
    ax.set_title(f'Step Response  (ζ={zeta:.2f},  ωₙ={omega_n:.1f})')
    ax.grid(alpha=0.3)
    ax.legend()
    plt.show()

zeta_slider = widgets.FloatSlider(value=0.5, min=0.05, max=1.5, step=0.05,
                                   description='Damping ζ')
wn_slider = widgets.FloatSlider(value=5.0, min=1.0, max=15.0, step=0.5,
                                 description='Natural freq ωₙ')

widgets.interact(plot_step, zeta=zeta_slider, omega_n=wn_slider);
```
