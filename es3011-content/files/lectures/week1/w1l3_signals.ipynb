# Signals: Continuous-Time and Discrete-Time

**Learning Goals**
- Understand what signals are in a control system (reference, error, control, output, disturbance).
- Understand the difference between continuous-time and discrete-time signals.

---

## What are signals in a control system?

In any control loop, several signals exist:

| Signal | Symbol | Description |
|--------|--------|-------------|
| Reference (setpoint) | $r(t)$ | Desired output value |
| Error | $e(t) = r(t) - y(t)$ | Difference between desired and actual |
| Control input | $u(t)$ | Command sent to the plant |
| Disturbance | $d(t)$ | Unwanted external input |
| Output | $y(t)$ | Actual measured value |

---

## Continuous and Discrete-Time Signals

### Continuous-Time Signals

A **continuous-time signal** $x(t)$ is defined for every instant of time $t$. Most physical quantities (position, velocity, voltage, current) are naturally continuous in time.

Examples:
- Position of a block: $x(t)$ for all $t \geq 0$
- Voltage across a capacitor: $v(t)$ for all $t \geq 0$
- Force applied to a system: $F(t)$ for all $t \geq 0$

In control engineering, we model continuous-time signals using differential equations and the Laplace transform.

---

### Discrete-Time Signals

A **discrete-time signal** $x[k]$ is defined only at specific instants of time, typically at integer multiples of a sampling period $T_s$:

$$x[k] = x(k T_s), \quad k = 0, 1, 2, \dots$$

Discrete-time signals arise naturally when a digital computer is part of the control loop. The computer can only read sensor values and update control commands at discrete sampling instants.

Examples:
- Sampled encoder readings in a digital motor controller
- GPS position updates at 10 Hz
- Digital audio samples at 44.1 kHz

In control engineering, we model discrete-time signals using difference equations and the $z$-transform.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

print("Libraries loaded.")
```

## Interactive: Continuous-Time Sinusoid

A continuous-time sinusoid:

$$x(t) = A \sin(2\pi f t + \phi)$$

Use the sliders to adjust amplitude $A$, frequency $f$, and phase $\phi$.

```python
def plot_continuous(A=1.0, freq=1.0, phase=0.0):
    t = np.linspace(0, 3, 3000)
    phi_rad = np.deg2rad(phase)
    x = A * np.sin(2 * np.pi * freq * t + phi_rad)
    fig, ax = plt.subplots(figsize=(8, 2.5))
    ax.plot(t, x, 'b-', lw=1.5)
    ax.axhline(0, color='k', lw=0.5)
    ax.set(xlabel='Time (s)', ylabel='$x(t)$',
           title=f'Continuous: $x(t) = {A:.1f} \\sin(2\\pi \\cdot {freq:.1f} t + {phase:.0f}^\\circ)$')
    ax.set_ylim(-1.5, 1.5)
    ax.grid(alpha=0.3)
    fig.tight_layout()
    plt.show()

w_A = widgets.FloatSlider(min=0.1, max=1.5, step=0.1, value=1.0,
                          description='$A$ (amplitude):',
                          style={'description_width': 'initial'})
w_f = widgets.FloatSlider(min=0.5, max=5.0, step=0.25, value=1.0,
                          description='$f$ (frequency, Hz):',
                          style={'description_width': 'initial'})
w_p = widgets.FloatSlider(min=0, max=360, step=15, value=0,
                          description='$\\phi$ (phase, deg):',
                          style={'description_width': 'initial'})
out_c = widgets.Output()

def update_c(change):
    with out_c:
        clear_output(wait=True)
        plot_continuous(w_A.value, w_f.value, w_p.value)

w_A.observe(update_c, 'value')
w_f.observe(update_c, 'value')
w_p.observe(update_c, 'value')
display(widgets.VBox([w_A, w_f, w_p, out_c]))
update_c(None)
```

## Interactive: Discrete-Time Sinusoid

A discrete-time sinusoid sampled at period $T_s$:

$$x[k] = A \sin(2\pi f \cdot k T_s + \phi), \qquad k = 0, 1, 2, \dots$$

The stars show the sampled values; the dashed line is the underlying continuous signal.

```python
def plot_discrete(A=1.0, freq=1.0, phase=0.0, Ts=0.1):
    phi_rad = np.deg2rad(phase)
    t_cont = np.linspace(0, 3, 3000)
    x_cont = A * np.sin(2 * np.pi * freq * t_cont + phi_rad)
    k = np.arange(0, int(3 / Ts) + 1)
    t_disc = k * Ts
    x_disc = A * np.sin(2 * np.pi * freq * t_disc + phi_rad)
    fig, ax = plt.subplots(figsize=(8, 2.5))
    ax.plot(t_cont, x_cont, 'b-', lw=1, alpha=0.4, label='Continuous')
    ax.stem(t_disc, x_disc, linefmt='r-', markerfmt='ro', basefmt='k-',
            label=f'Sampled ($T_s = {Ts:.2f}$ s)')
    ax.set(xlabel='Time (s)', ylabel='$x[k]$',
           title=f'Discrete: $A = {A:.1f},\\; f = {freq:.1f}\\, \\text{{Hz}},\\; T_s = {Ts:.2f}\\, \\text{{s}}$')
    ax.set_ylim(-1.5, 1.5)
    ax.legend(fontsize=9)
    ax.grid(alpha=0.3)
    fig.tight_layout()
    plt.show()

wd_A = widgets.FloatSlider(min=0.1, max=1.5, step=0.1, value=1.0,
                           description='$A$ (amplitude):',
                           style={'description_width': 'initial'})
wd_f = widgets.FloatSlider(min=0.5, max=5.0, step=0.25, value=1.0,
                           description='$f$ (frequency, Hz):',
                           style={'description_width': 'initial'})
wd_p = widgets.FloatSlider(min=0, max=360, step=15, value=0,
                           description='$\\phi$ (phase, deg):',
                           style={'description_width': 'initial'})
wd_Ts = widgets.FloatSlider(min=0.02, max=0.5, step=0.02, value=0.1,
                            description='$T_s$ (sampling period, s):',
                            style={'description_width': 'initial'})
out_d = widgets.Output()

def update_d(change):
    with out_d:
        clear_output(wait=True)
        plot_discrete(wd_A.value, wd_f.value, wd_p.value, wd_Ts.value)

wd_A.observe(update_d, 'value')
wd_f.observe(update_d, 'value')
wd_p.observe(update_d, 'value')
wd_Ts.observe(update_d, 'value')
display(widgets.VBox([wd_A, wd_f, wd_p, wd_Ts, out_d]))
update_d(None)
```
