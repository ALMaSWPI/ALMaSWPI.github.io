## The Three Mechanical Elements

Every mechanical system can be decomposed into three ideal components:

| Element | Symbol | Law | Units |
|---|---|---|---|
| **Mass** (inertia) | $m$ | $F = m\,\ddot x$ | kg |
| **Spring** (stiffness) | $k$ | $F = k\,x$ | N/m |
| **Damper** (viscous friction) | $b$ | $F = b\,\dot x$ | N·s/m |

The differential equation emerges from **force balance** (Newton's second law) applied to a free-body diagram.

---

### Mass

A mass stores kinetic energy. When a net force acts on a mass, it **accelerates**:

$$\sum F = m\,\ddot x$$

This is Newton's second law. In the ODE, the mass produces the **second-order** term $m\ddot x(t)$.

---

### Spring

A spring stores potential energy. The force it exerts is proportional to **displacement** (Hooke's law):

$$F_{\text{spring}} = k\,x$$

The spring always acts to **restore** the mass to its equilibrium position. In the ODE it contributes the $k\,x(t)$ term. The sign is opposite to the displacement.

---

### Damper

A damper (dashpot) dissipates energy. The force it exerts is proportional to **velocity** (viscous friction):

$$F_{\text{damper}} = b\,\dot x$$

The damper opposes motion, as it always acts against the direction of travel. In the ODE it contributes the $b\,\dot x(t)$ term. Without a damper, the system would oscillate forever once disturbed.

---

## Putting it Together

Consider a mass $m$ connected to a rigid wall through a spring $k$ and a damper $b$, with an external force $F(t)$ applied to the mass.

The free-body diagram gives:

$$m\,\ddot x = F(t) - b\,\dot x - k\,x$$

Rearranged into the standard form:

$$m\,\ddot x(t) + b\,\dot x(t) + k\,x(t) = F(t)$$

Each element maps to one piece of the differential equation.

---

## Summary

- **Mass** → $m\ddot x$ stores kinetic energy, gives second-order behaviour.
- **Spring** → $k x$ stores potential energy, provides restoring force.
- **Damper** → $b\dot x$ dissipates energy, provides damping.
- The full equation $m\ddot x + b\dot x + k x = F(t)$ is a direct sum of the three component laws via Newton's second law.
- Dividing through by $m$ gives the standard form $\ddot x + 2\zeta\omega_n\dot x + \omega_n^2 x = \frac{1}{m}F(t)$, where $\omega_n = \sqrt{k/m}$ and $\zeta = b/(2\sqrt{km})$.
