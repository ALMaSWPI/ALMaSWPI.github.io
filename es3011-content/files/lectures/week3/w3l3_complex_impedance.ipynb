## Complex Impedance Definition

Complex impedance $Z$ generalizes resistance to the frequency domain. Substitute $s = j\omega$:

| Component | Time domain | Laplace domain $Z(s)$ | Complex impedance $Z(j\omega)$ |
|-----------|------------|----------------------|-------------------------------|
| **Resistor** | $v = iR$ | $Z_R(s) = R$ | $Z_R = R$ |
| **Inductor** | $v = L\frac{di}{dt}$ | $Z_L(s) = Ls$ | $Z_L = j\omega L$ |
| **Capacitor** | $i = C\frac{dv}{dt}$ | $Z_C(s) = \dfrac{1}{Cs}$ | $Z_C = \dfrac{1}{j\omega C} = -\dfrac{j}{\omega C}$ |

**Frequency-dependent behavior:**

| $\omega$ | Inductor $Z_L = j\omega L$ | Capacitor $Z_C = 1/j\omega C$ |
|----------|---------------------------|------------------------------|
| DC ($\omega = 0$) | Short circuit ($Z_L = 0$) | Open circuit ($Z_C \to \infty$) |
| High ($\omega \to \infty$) | Open circuit ($Z_L \to \infty$) | Short circuit ($Z_C \to 0$) |

---
## Impedance Reduction Rules

Just like resistors, complex impedances combine in series and parallel using the same formulas:

**Series:**
$$Z_{\text{eq}} = Z_1 + Z_2 + \dots + Z_n$$

**Parallel:**
$$\frac{1}{Z_{\text{eq}}} = \frac{1}{Z_1} + \frac{1}{Z_2} + \dots + \frac{1}{Z_n}$$

For **two components in parallel:**
$$Z_{\text{eq}} = \frac{Z_1 \, Z_2}{Z_1 + Z_2}$$

---
### Example 1: Two Resistors in Series

For resistors, impedance is purely real: $Z_R = R$. The series equivalent is simply the sum:

$$Z_{\text{eq}} = R_1 + R_2$$

```python
R1_s = widgets.FloatSlider(min=1, max=100, step=1, value=10,
                           description='R1 (\u03A9):',
                           style={'description_width': 'initial'})
R2_s = widgets.FloatSlider(min=1, max=100, step=1, value=20,
                           description='R2 (\u03A9):',
                           style={'description_width': 'initial'})
run_r = widgets.Button(description='Run', button_style='primary')
out_r = widgets.Output()

def on_run_r(b):
    R1 = R1_s.value
    R2 = R2_s.value
    Z_eq = R1 + R2
    with out_r:
        clear_output(wait=True)
        with schemdraw.Drawing() as d:
            elm.SourceV().label('Vin').up()
            elm.Resistor().label(f'R1={R1:.0f}\u03A9').right()
            elm.Resistor().label(f'R2={R2:.0f}\u03A9').right()
            elm.Line().down()
            elm.Ground()
        print(f'Series impedance:')
        print(f'  Z_eq = R1 + R2 = {R1:.0f} + {R2:.0f} = {Z_eq:.0f} \u03A9')
        print(f'  (purely real, no frequency dependence)')

run_r.on_click(on_run_r)
display(widgets.VBox([R1_s, R2_s, run_r, out_r]))
print("Adjust R1 and R2, then click Run.")
```

### Example 2: Series RLC Circuit

For a resistor $R$, inductor $L$, and capacitor $C$ in series, the total impedance is the sum:

$$Z_{\text{eq}} = Z_R + Z_L + Z_C = R + j\omega L + \frac{1}{j\omega C}$$

This can be rewritten as:

$$Z_{\text{eq}} = R + j\left(\omega L - \frac{1}{\omega C}\right) = |Z| \, e^{j\phi}$$

where the magnitude and phase are:

$$|Z| = \sqrt{R^2 + \left(\omega L - \frac{1}{\omega C}\right)^2}, \qquad
\phi = \tan^{-1}\left(\frac{\omega L - 1/\omega C}{R}\right)$$

```python
R_s = widgets.FloatSlider(min=1, max=100, step=1, value=10,
                          description='R (\u03A9):',
                          style={'description_width': 'initial'})
L_s = widgets.FloatSlider(min=0.1, max=10, step=0.1, value=1,
                          description='L (H):',
                          style={'description_width': 'initial'})
C_s = widgets.FloatSlider(min=0.1e-6, max=10e-6, step=0.1e-6, value=1e-6,
                          description='C (F):',
                          readout_format='.2e',
                          style={'description_width': 'initial'})
w_s = widgets.FloatSlider(min=1, max=10000, step=10, value=1000,
                          description='\u03C9 (rad/s):',
                          style={'description_width': 'initial'})
run_rlc = widgets.Button(description='Run', button_style='primary')
out_rlc = widgets.Output()

def on_run_rlc(b):
    R = R_s.value
    L = L_s.value
    C = C_s.value
    w = w_s.value
    Z = R + 1j * w * L + 1 / (1j * w * C)
    Zmag = abs(Z)
    Zph = np.angle(Z, deg=True)
    with out_rlc:
        clear_output(wait=True)
        with schemdraw.Drawing() as d:
            elm.SourceV().label('Vin').up()
            elm.Resistor().label(f'R={R:.0f}\u03A9').right()
            elm.Inductor().label(f'L={L:.1f}H').right()
            elm.Capacitor().label(f'C={C*1e6:.1f}\u03BCF').right()
            elm.Line().down()
            elm.Ground()
        print(f'Components: R={R}\u03A9, L={L}H, C={C*1e6:.1f}\u03BCF')
        print(f'Frequency: \u03C9 = {w} rad/s')
        print()
        print(f'Inductive reactance:  XL = j\u03C9L = j{w*L:.2f} \u03A9')
        print(f'Capacitive reactance: XC = 1/(j\u03C9C) = -j{1/(w*C):.2f} \u03A9')
        print()
        print(f'Z_eq = R + jXL + XC')
        print(f'     = {R:.2f} + j{w*L:.2f} - j{1/(w*C):.2f}')
        print(f'     = {Z.real:.2f} + j({Z.imag:.2f})')
        print()
        print(f'|Z|  = {Zmag:.2f} \u03A9')
        print(f'\u03d5   = {Zph:.1f} deg')

run_rlc.on_click(on_run_rlc)
display(widgets.VBox([R_s, L_s, C_s, w_s, run_rlc, out_rlc]))
print("Adjust values, then click Run.")
```

---
## Summary

| Component | Impedance $Z(j\omega)$ |
|-----------|----------------------|
| Resistor | $Z_R = R$ |
| Inductor | $Z_L = j\omega L$ |
| Capacitor | $Z_C = 1/j\omega C = -j/(\omega C)$ |

**Reduction rules** (same as resistors):
- Series: $Z_{\text{eq}} = Z_1 + Z_2 + \dots$
- Parallel: $1/Z_{\text{eq}} = 1/Z_1 + 1/Z_2 + \dots$
- Two in parallel: $Z_{\text{eq}} = Z_1 Z_2 / (Z_1 + Z_2)$

**Key takeaway:** equivalent impedance is a complex number $Z = R + jX$. Its real part is resistance, its imaginary part is reactance. The reactance determines whether the circuit is inductive ($X > 0$) or capacitive ($X < 0$).
